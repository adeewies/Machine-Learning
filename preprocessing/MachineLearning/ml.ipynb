```python
from sklearn.datasets import make_regression
```

```python
X,y = make_regression(n_samples=8, n_features=4)
from sklearn.preprocessing import PolynomialFeatures
poly_transform = PolynomialFeatures(degree = 4, interaction_only=True)
X_trans = poly_transform.fit_transform(X,y)
print(X_trans.shape)
```
